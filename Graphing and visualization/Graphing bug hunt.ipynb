```python
x = sym.symbols('x')
y = x**2 - 3*x

xrange = range(-10,11)

for i in range(0,len(xrange)):
    plt.plot(xrange[i],y.subs(x,xrange[i]),'o')

plt.xlabel('x')
plt.ylabel('$f(x) = %s$' %sym.latex(y))
plt.show()
```

#populate the matrix

for i in range(0,m):

    for j in range(0,n):

```python
m = 8
n = 4

C = np.zeros((m,4))


for i in range(0,m):
    for j in range(0,n):
        C[i,j] = (-1)**(i+j)

for i in range(0,m):
     for j in range(0,n):
         plt.text(j,i,i+j,\
                  horizontalalignment='center',verticalalignment='center',\
                    fontdict=dict(color='m'))     

plt.imshow(C)
plt.set_cmap('gray')

plt.show()
```
