f(x) = -x^2

```python
#1)

y = np.array([ [-0.5,-2.5],[-0.5,-4],[0.5,-4],[0.5,-2.5] ])
p = Polygon(y,facecolor = 'k', alpha = 1)
 
x1 = np.linspace(-2,2,20)
y1 = [[-2,-4]]
 
for x in x1:
    temp = [[x,-x**2]]
    y1 = np.append(y1,temp,axis=0)
    
p1 = Polygon(y1,facecolor = 'b', alpha = .4 , edgecolor = 'k')
 
fig,ax = plt.subplots()
 
 
ax.add_patch(p1)
ax.add_patch(p)
 
ax.set_xlim(-2.3,2.3)
ax.set_ylim(-4.2,.2)
plt.show()
```

```python
#2)

x = np.linspace(-2,2,50)
y = -(x**2)
z = np.stack((x,y),axis=1)
z[0:4]

p1 = Polygon(z,facecolor='b',closed=True,alpha=0.2,edgecolor='k')
# The door
p2 = Polygon([[-0.5,-4],[-0.5,-2.5],[0.5,-2.5],[0.5,-4]],facecolor='k')
# The edge
p3 = Polygon(z,facecolor='none',edgecolor='k',linewidth=2 )

ig,ax = plt.subplots()
ax.add_patch(p1)
ax.add_patch(p2)
ax.add_patch(p3)

ax.set_ylim([-4.2,0.2])
ax.set_xlim([-2.2,2.2])
plt.show()
```

```python
#3)

X_coord=np.arange(-2,2.01,0.01)
y=-X_coord**2
A=np.array([[-0.5,-4],[-0.5,-2.5],[0.5,-4]])
B=np.array([[0.5,-4],[0.5,-2.5],[-0.5,-2.5]])
p1=Polygon(A,facecolor='k',edgecolor='k')
p2=Polygon(B,facecolor='k')
fig,ax=plt.subplots()
for i in range(0,401):
    p3=Polygon([  [X_coord[0],y[0]],  [X_coord[i],y[i]],  [X_coord[i],y[0]]],facecolor='k',edgecolor='k',alpha=1)
    ax.add_patch(p3)
for i in range(0,401):
    p4=Polygon([  [X_coord[0],y[0]],  [X_coord[-i],y[-i]],  [X_coord[-i],y[0]]],facecolor='c',edgecolor='k',alpha=0.01)
    ax.add_patch(p4)
ax.add_patch(p1)
ax.add_patch(p2)
 
 
    
ax.set_ylim([-4.1,0.1])
ax.set_xlim([-2.1,2.1])
plt.show()
```
