```python
x = sym.symbols('x')
y = x**2 - 3*x

for pnti in range(-11,12):
    plt.plot(pnti,y.subs(x,pnti),'o')

plt.xlabel('x')
plt.ylabel('f(x) = %s'%y)
plt.show()
```
