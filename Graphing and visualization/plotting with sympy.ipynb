```python
x,a = sym.symbols('x,a')

expr = a/x

n=1

p = symplot(expr.subs(a,1),(x,-5,5),show=False)
p[0].label = '$y=%s$' %sym.latex(expr.subs(a,n))

p2 = symplot(expr.subs(a,2),show=False)
p.extend(p2)
p[1].label = '$y=%s$'%sym.latex(expr.subs(a,2))

p[0].line_color = 'y'
p.ylim = [-5,5]
p.legend = True
p.show()
```
